# **<p style="font-family:newtimeroman;font-size:150%;text-align:center;color:#06445e;">Изменение типов данных столбцов с Object на Float</p>**

```python
# Изменение типа данных цены со скидкой и фактической цены

df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')
```

```python
# Изменение типа данных и значений в проценте скидки

df['discount_percentage'] = df['discount_percentage'].str.replace('%','').astype('float64')

df['discount_percentage'] = df['discount_percentage'] / 100
```

```python
# Нахождение необычной строки в колонке рейтинга
df['rating'].value_counts()
```

```python
# Проверка странной колонки
df.query('rating == "|"')
```

### ***`Рейтинг` - 3,9. Поэтому я собираюсь дать `рейтинг` элементу также 3,9.***

```python
# Изменение типа данных столбцов рейтинга

df['rating'] = df['rating'].str.replace('|', '3.9').astype('float64')
```

```python
# Изменение типа данных столбца «rating_count»

df['rating_count'] = df['rating_count'].str.replace(',', '').astype('float64')
```

```python
from sklearn.preprocessing import LabelEncoder
# метка кодирует категориальные переменные

le_product_id = LabelEncoder()
le_category = LabelEncoder()
le_review_id = LabelEncoder()
le_review_content = LabelEncoder()
le_product_name = LabelEncoder()
le_user_name = LabelEncoder()
le_about_product = LabelEncoder()
le_user_id = LabelEncoder()
le_review_title = LabelEncoder()
le_img_link = LabelEncoder()
le_product_link = LabelEncoder()


df['product_id'] = le_product_id.fit_transform(df['product_id'])
df['category'] = le_category.fit_transform(df['category'])
df['review_id'] = le_review_id.fit_transform(df['review_id'])
df['review_content'] = le_review_content.fit_transform(df['review_content'])
df['product_name'] = le_product_name.fit_transform(df['product_name'])
df['user_name'] = le_user_name.fit_transform(df['user_name'])
df['about_product'] = le_about_product.fit_transform(df['about_product'])
df['user_id'] = le_user_id.fit_transform(df['user_id'])
df['review_title'] = le_review_title.fit_transform(df['review_title'])
df['img_link'] = le_img_link.fit_transform(df['img_link'])
df['product_link'] = le_product_link.fit_transform(df['product_link'])
```

```python
# Рассчитываем коэффициенты корреляции Пирсона (по умолчанию в Pandas)
correlation_matrix = df.corr()

# Выводим матрицу корреляции
print(correlation_matrix)

# Создаем тепловую карту для визуализации корреляций
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Pearson)")
plt.show()

# Рассчитываем коэффициенты корреляции Спирмена (для нелинейных отношений)
spearman_correlation_matrix = df.corr(method="spearman")

# Выводим корреляционную матрицу Спирмена
print(spearman_correlation_matrix)

# Создаем тепловую карту для визуализации корреляций Спирмена
sns.heatmap(spearman_correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix (Spearman)")
plt.show()
```

# <p style="font-family:newtimeroman;font-size:150%;text-align:center;color:#06445e;">Статистические тесты:</p>

```python
import scipy.stats as stats

# Проводим t-тест для сравнения рейтинга между двумя категориями
t_statistic, p_value = stats.ttest_ind(df[df['category'] == 'electronics']['rating'], df[df['category'] == 'clothing']['rating'])

# Печатаем t-статистику и p-значение
print(t_statistic, p_value)
```

```python
df.info()
```

```python
# Тест хи-квадрат

# Создаем таблицу непредвиденных обстоятельств
contigency_table = pd.crosstab(df['actual_price'], df['rating'])
contigency_table
```

```python
# выполняем хи-квадрат тест
chi2, p, dof, expected = stats.chi2_contingency(contigency_table)

# print the results
print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print(f"Expected:\n {expected}")
```

## ***Q6: Какие ключевые слова продуктов самые популярные?*** 

```python
def extract_keywords(product_name):
  """Извлекает ключевые слова из названия продукта, обрабатывая потенциальные числа."""
  if isinstance(product_name, str):  # Проверьте, является ли это строкой
    keywords = product_name.lower().split()  # Разделяет на слова и строчные буквы
    keywords = [word for word in keywords if word.isalpha()]  # Удаляет неалфавитные символы
  else:
    keywords = []  # Обрабатывает нестроковые значения (например, целые числа) путем возврата пустого списка
  return keywords

# Применяем функцию для извлечения ключевых слов
df["keywords"] = df["product_name"].apply(extract_keywords)

# Сглаживаем список ключевых слов
all_keywords = [keyword for keywords in df["keywords"] for keyword in keywords]

# Count keyword occurrences
keyword_counts = pd.Series(all_keywords).value_counts()

# Display the top 10 most popular keywords
print(keyword_counts.head(10))
```

### ***Ответ 6:***
- ***`USB`соединение, `зарядка (особенно быстрая)` и `кабели` являются важными характеристиками продукта.***
- ***Предлоги и союзы ``с``, ``для``, ``и``, ``для``, ``для``, ``для`` свидетельствуют о том, что основное внимание уделяется объяснению совместимости продуктов и сценариям их использования.***
- ***`Кабели и интеллектуальные устройства`, вероятно, хорошо представлены в наборе данных.***
- ***`Названия продуктов`, как правило, лаконичны и используют «общеупотребительные слова», что потенциально может быть полезно при использовании усовершенствованных методов извлечения ключевых слов.***

#### - ***`Присутствие "Простых калькуляторов"` в топ-5 указывает на то, что даже относительно простые продукты могут `получать высокие рейтинги`, если они эффективно удовлетворяют потребности клиентов.***
